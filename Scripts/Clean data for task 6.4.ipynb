```python
def rename_columns(df, color_prefix):
    # Create a new dictionary to rename columns dynamically
    new_columns = {}
    
    for col in df.columns:
        if col.startswith(color_prefix):
            # Remove the color prefix and replace underscores with spaces
            new_name = col[len(color_prefix):].replace('_', ' ')
            new_columns[col] = new_name
    
    # Rename the columns in the dataframe
    df = df.rename(columns=new_columns)
    
    return df

# Apply the renaming function to both red and blue winners dataframes
df_red_winners = rename_columns(df_red_winners, 'R_')
df_blue_winners = rename_columns(df_blue_winners, 'B_')

# Check the renamed columns
print("Renamed columns in Red Winners DataFrame:")
print(df_red_winners.columns.tolist())

print("\nRenamed columns in Blue Winners DataFrame:")
print(df_blue_winners.columns.tolist())
```

```python
def format_column_name(col_name):
    # Replace underscores with spaces and capitalize each word
    return ' '.join(word.capitalize() for word in col_name.split('_'))

def rename_columns(df, color_prefix):
    # Create a new dictionary to rename columns dynamically
    new_columns = {}
    
    for col in df.columns:
        if col.startswith(color_prefix):
            # Remove the color prefix and format the column name
            new_name = format_column_name(col[len(color_prefix):])
            new_columns[col] = new_name
    
    # Rename the columns in the dataframe
    df = df.rename(columns=new_columns)
    
    return df

# Apply the renaming function to both red and blue winners dataframes
df_red_winners = rename_columns(df_red_winners, 'R_')
df_blue_winners = rename_columns(df_blue_winners, 'B_')

# Check the renamed columns
print("Renamed columns in Red Winners DataFrame:")
print(df_red_winners.columns.tolist())

print("\nRenamed columns in Blue Winners DataFrame:")
print(df_blue_winners.columns.tolist())
```

```python
def move_columns(df):
    # Get the list of columns
    cols = df.columns.tolist()
    
    # Identify the columns to move
    move_cols = ['City', 'State', 'Country']
    
    # Find the index of 'Fight Location'
    loc_index = cols.index('Fight Location')
    
    # Remove the columns to move from their original positions
    for col in move_cols:
        cols.remove(col)
    
    # Insert the columns after 'Fight Location'
    for i, col in enumerate(move_cols):
        cols.insert(loc_index + 1 + i, col)
    
    # Reorder the dataframe
    df = df[cols]
    
    return df

# Apply the function to both red and blue winners dataframes
df_red_winners = move_columns(df_red_winners)
df_blue_winners = move_columns(df_blue_winners)

# Check the columns order
print("Columns in Red Winners DataFrame:")
print(df_red_winners.columns.tolist())

print("\nColumns in Blue Winners DataFrame:")
print(df_blue_winners.columns.tolist())
```

```python
def capitalize_columns(df):
    # Create a new dictionary with capitalized column names
    new_columns = {col: col.capitalize() for col in df.columns}
    
    # Rename the columns in the dataframe
    df = df.rename(columns=new_columns)
    
    return df

# Apply the function to both red and blue winners dataframes
df_red_winners = capitalize_columns(df_red_winners)
df_blue_winners = capitalize_columns(df_blue_winners)

# Check the renamed columns
print("Capitalized columns in Red Winners DataFrame:")
print(df_red_winners.columns.tolist())

print("\nCapitalized columns in Blue Winners DataFrame:")
print(df_blue_winners.columns.tolist())
```
