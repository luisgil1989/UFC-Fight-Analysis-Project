```python
df['Winner'] = 1
# Ensure 'Winner' is correctly encoded as integer
df['Winner'] = df['Winner'].astype(int)

# Filter numeric columns only
numeric_df = df.select_dtypes(include=[int, float])
```

```python
# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()
```

```python
# Define variable groups with "Wins" as the target variable
striking_vars = [
    'Wins', 
    'Average Significant Strikes Landed', 'Average Significant Strikes Attempted',
    'Average Total Strikes Landed', 'Average Total Strikes Attempted',
    'Average Head Strikes Landed', 'Average Head Strikes Attempted',
    'Average Body Strikes Landed', 'Average Body Strikes Attempted',
    'Average Leg Strikes Landed', 'Average Leg Strikes Attempted'
]

grappling_vars = [
    'Wins', 
    'Average Takedowns Landed', 'Average Takedown Attempts',
    'Average Submission Attempts', 'Average Reversals',
    'Average Clinch Strikes Landed', 'Average Clinch Strikes Attempted',
    'Average Ground Strikes Landed', 'Average Ground Strikes Attempted',
    'Average Control Time (Seconds)', 'Average Control Time Taken (Seconds)'
]

outcome_vars = [
    'Wins', 
    'Current Win Streak', 'Longest Win Streak',
    'Wins By Decision Majority', 'Wins By Decision Split', 'Wins By Decision Unanimous',
    'Wins By KO/TKO', 'Wins By Submission', 'Wins By TKO Doctor Stoppage'
]
```

# Plotting the correlation maps

```python
# Function to plot correlation heatmap
def plot_correlation_heatmap(df, variables, title):
    correlation_matrix = df[variables].corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(title)
    plt.show()
```

```python
plot_correlation_heatmap(df, striking_vars, 'Correlation Map: Striking Metrics + Wins')
```

```python
plot_correlation_heatmap(df, grappling_vars, 'Correlation Map: Grappling Metrics + Wins')
```

```python
plot_correlation_heatmap(df, outcome_vars, 'Correlation Map: Fight Outcome Metrics + Wins')
```

### Which Metric Most Predicts Fight Outcomes?

Based on the correlation analysis and the interpretation of the correlation maps, the metrics that most predict fight outcomes, particularly in terms of total wins, can be identified by looking at the strongest correlations with the "Wins" column.

#### Summary of Key Findings:

1. **Longest Win Streak (Correlation: 0.86)**
   - **Description**: This metric represents the longest consecutive winning streak a fighter has achieved in their career.
   - **Significance**: A high correlation with "Wins" indicates that fighters who have long winning streaks tend to accumulate more wins overall. This suggests that consistency in winning is a strong predictor of overall success.

2. **Wins By Decision Unanimous (Correlation: 0.78)**
   - **Description**: This metric represents the number of fights a fighter has won by unanimous decision.
   - **Significance**: The strong correlation with "Wins" shows that fighters who frequently win by unanimous decision tend to have a higher number of total wins. This indicates that the ability to convincingly win on the judges' scorecards is a key factor in accumulating wins.

3. **Wins By KO/TKO (Correlation: 0.75)**
   - **Description**: This metric counts the number of wins a fighter has achieved by knockout or technical knockout.
   - **Significance**: The strong correlation suggests that fighters who are successful in achieving knockouts or technical knockouts tend to have more total wins. This highlights the importance of striking power in fight outcomes.

#### Conclusion:
- **Most Predictive Metric**: **Longest Win Streak** is the metric that most predicts fight outcomes in terms of total wins, given its highest correlation with the "Wins" column (0.86).
- **Other Significant Metrics**: **Wins By Decision Unanimous** and **Wins By KO/TKO** also strongly predict total wins and indicate that both consistent performance in decision victories and the ability to finish fights by KO/TKO are important.
